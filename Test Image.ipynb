```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
import os
import json
import cv2
import numpy as np
```

```python
image = cv2.imread('../train_val_images/Mammalia/Marmota flaviventris/6c328abfae05cc8820e884d4fcafebf3.jpg')
```

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
with open('train_2017_bboxes.json') as json_file:
    data = json.load(json_file)
    
    
# for i, data in enumerate(data['annotations']):
    
print(data['annotations'][94])

lt_x, lt_y, width, height = data['annotations'][94]['bbox']

rb_x, rb_y = lt_x + width, lt_y + height

print(lt_x, lt_y, rb_x, rb_y)
```

```python
cv2.rectangle(RGB_img,(lt_x, lt_y), (rb_x, rb_y), (255, 0, 0), 2)
plt.imshow(RGB_img)
```
